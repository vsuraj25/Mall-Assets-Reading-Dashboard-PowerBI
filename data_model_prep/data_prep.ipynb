### Power Consumption

```python
def power_consumption(run_status, type, space):
    power_c_list = []
    for i in range(len(type)):
        if run_status[i] == 'ON':
            if type[i] == 'Window AC' and space[i] == 'Small':
                power_c_list.append(round(random.uniform(0.5,0.8),1))
            elif type[i] == 'Window AC' and space[i] == 'Medium':
                power_c_list.append(round(random.uniform(0.9,1.5),1))
            elif type[i] == 'Window AC' and space[i] == 'Large':
                power_c_list.append(round(random.uniform(1.6,2.5),1))
            elif type[i] == 'Split System' and space[i] == 'Small':
                power_c_list.append(round(random.uniform(0.5,0.9),1))
            elif type[i] == 'Split System' and space[i] == 'Medium':
                power_c_list.append(round(random.uniform(1.0,1.7),1))
            elif type[i] == 'Split System' and space[i] == 'Large':
                power_c_list.append(round(random.uniform(1.8,2.5),1))
        else:
            power_c_list.append(0)

    return power_c_list
```

```python
power_consumption_list = power_consumption(data['Run_Status'], data['Type'], data['Space'])
```

```python
data['Power_Consumption_in_kwh'] = power_consumption_list
```

### Airflow Rate in CFM

```python
def airflow_rate(run_status, type, space):
    airflow_rate = []
    for i in range(len(type)):
        if run_status[i] == 'ON':
            if type[i] == 'Window AC' and space[i] == 'Small':
                airflow_rate.append(random.randint(100,250))
            elif type[i] == 'Window AC' and space[i] == 'Medium':
                airflow_rate.append(random.randint(251, 400))
            elif type[i] == 'Window AC' and space[i] == 'Large':
                airflow_rate.append(random.randint(400, 500))
            elif type[i] == 'Split System' and space[i] == 'Small':
                airflow_rate.append(random.randint(100, 300))
            elif type[i] == 'Split System' and space[i] == 'Medium':
                airflow_rate.append(random.randint(300, 400))
            elif type[i] == 'Split System' and space[i] == 'Large':
                airflow_rate.append(random.randint(400, 500))

        else:
            airflow_rate.append(0)
    return airflow_rate
```

```python
airflow_rates = airflow_rate(data['Run_Status'], data['Type'], data['Space'])
```

```python
data['Airflow_Rates_in_CFM'] = airflow_rates
```
