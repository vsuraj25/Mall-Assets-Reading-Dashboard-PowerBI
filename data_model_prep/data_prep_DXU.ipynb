### Power Consumption

```python
def power_consumption(run_status, type, space):
    power_c_list = []
    for i in range(len(type)):
        if run_status[i] == 'ON':
            if type[i] == 'Window AC' and space[i] == 'Small':
                power_c_list.append(round(random.uniform(0.5,0.8),1))
            elif type[i] == 'Window AC' and space[i] == 'Medium':
                power_c_list.append(round(random.uniform(0.9,1.5),1))
            elif type[i] == 'Window AC' and space[i] == 'Large':
                power_c_list.append(round(random.uniform(1.6,2.5),1))
            elif type[i] == 'Split System' and space[i] == 'Small':
                power_c_list.append(round(random.uniform(0.5,0.9),1))
            elif type[i] == 'Split System' and space[i] == 'Medium':
                power_c_list.append(round(random.uniform(1.0,1.7),1))
            elif type[i] == 'Split System' and space[i] == 'Large':
                power_c_list.append(round(random.uniform(1.8,2.5),1))
        else:
            power_c_list.append(0)

    return power_c_list
```

```python
power_consumption_list = power_consumption(data['Run_Status'], data['Type'], data['Space'])
```

```python
data['Power_Consumption_in_kwh'] = power_consumption_list
```

### Airflow Rate in CFM

```python
def airflow_rate(run_status, type, space):
    airflow_rate = []
    for i in range(len(type)):
        if run_status[i] == 'ON':
            if type[i] == 'Window AC' and space[i] == 'Small':
                airflow_rate.append(random.randint(100,250))
            elif type[i] == 'Window AC' and space[i] == 'Medium':
                airflow_rate.append(random.randint(251, 400))
            elif type[i] == 'Window AC' and space[i] == 'Large':
                airflow_rate.append(random.randint(400, 500))
            elif type[i] == 'Split System' and space[i] == 'Small':
                airflow_rate.append(random.randint(100, 300))
            elif type[i] == 'Split System' and space[i] == 'Medium':
                airflow_rate.append(random.randint(300, 400))
            elif type[i] == 'Split System' and space[i] == 'Large':
                airflow_rate.append(random.randint(400, 500))

        else:
            airflow_rate.append(0)
    return airflow_rate
```

```python
airflow_rates = airflow_rate(data['Run_Status'], data['Type'], data['Space'])
```

```python
data['Airflow_Rates_in_CFM'] = airflow_rates
```

```python
def create_data_dxunit():

    def trip_status(run_status):
        if run_status == 'On':
            return random.choice(['Normal', 'Trip'], p = [0.8, 0.2])
        else:
            return 'Normal'

    def filter_schedule(run_status):
        if run_status == "On":
            return random.choice(['Off', 'Scheduled'])
        else:
            return 'Off'
        
    def power_consumption(run_status):
            if run_status == 'On':
                return round(random.uniform(1.8,2.5),1)
            else:
                return 0
    
    def airflow_rate(run_status):
        if run_status == 'On':
                return random.randint(350, 500)
        else:
            return 0
    
    locations = {'Mumbai': ['Paladium', 'Marketcity'], 'Bangalore' : ['Marketcity'], 'Pune': ['Marketcity'], 'Chennai': ['Paladium', 'Marketcity'], 'Lucknow': ['Palassio', 'United'], 'Bareilly': ['United']}
    for loc in locations.keys():
        for mall in locations[loc]:

            datelist = pd.date_range(start='2021-01-01', end = '2022-12-31').tolist()
            data = pd.DataFrame(datelist, columns=['Date'])
            data_size = data.shape[0]
            
            ### Location
            data['Location'] = loc

            ### Mall
            data['Mall'] =  mall

            ### System ON and OFF Command
            data['On_Off_Cmd'] = 'On'

            ### Mode(Auto/Manual) Status
            data['A/M_Status'] = random.choice(['Auto', 'Manual'], p = [0.6, 0.4], size=data_size)

            ### Run Status
            data['Run_Status'] = random.choice(['On', 'Off'], p = [0.95, 0.05], size=data_size)
            
            ### Trip Status
            data['Trip_Status'] = data['Run_Status'].apply(trip_status)
            
            ### Schedule Mode
            data['Schedule_Mode'] = data['Run_Status'].apply(filter_schedule)
            
            ### Time Schedule
            # If the system is scheduled, it can be only scheduled based on time
            data['Time_Schedule'] = data['Schedule_Mode']

            ### Power Consumption
            data['Power_Consumption_in_kwh'] = data['Run_Status'].apply(power_consumption)
            
            ### Filter Status
            data['Filter_Status'] = random.choice(['Clean', 'Dirty'], p = [0.8, 0.2], size=data_size)
            
            ### Filter Status
            data['Indoor_Temperature'] = random.randint(15, 28, size=data_size)
            
            ### Airflow Rate in CFM
            data['Airflow_Rates_in_CFM'] = data['Run_Status'].apply(airflow_rate)
            
            data.to_csv(f'data/DX_Unit/{loc}_{mall}_Assets.csv', index = False)

    
```
